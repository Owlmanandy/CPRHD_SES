```python
import pandas as pd
import os
# dp03 is econ data, dp02 is social, dp05 is race
```

```python
def combine_f(x,y,z,year):
    x.columns = x.iloc[0]
    y.columns = y.iloc[0]
    y = y.drop(0)
    x = x.drop(0)
    z.columns = z.iloc[0]
    z = z.drop(0)
    combined = z.merge(x.merge(y,how = 'inner',on = 'Id'),how = 'inner', on = 'Id')
    return combined
```

```python
directory = os.curdir
li = []
year_list = [10,11,12,13,14,15,16]
for d in os.listdir(directory):
    if d.startswith("2"):
        for files in os.listdir(d) :
            if files.endswith('.csv'):
                year = d[:4]
                path = os.path.join(directory, d)
                file_path = os.path.join(path, files)
                if('DP03' in files):
                    econ = pd.read_csv(file_path)
                elif('DP02' in files):
                    social = pd.read_csv(file_path)
                elif('DP05' in files):
                    race = pd.read_csv(file_path)
        combine = combine_f(econ,social,race,year)
        combine.to_csv(year + ' combined.csv')
    else:
        continue
```

```python
# clean all combine files
def clean(x):
    x['Geography'] = x['Geography'].apply(lambda x: x.split()[2][:-1])
    x = x.set_index('Geography').drop(delete_list)
    return x.groupby(x.index).first()
```
